```python
fig, ax = plt.subplots()
salaries.plot(kind="box")

def formatter(x, pos):
    return str(round(x / 1e5, 2)) + " lakh"


ax.yaxis.set_major_formatter(formatter)
ax.yaxis.set_minor_formatter(NullFormatter())
ax.yaxis.set_major_locator(ticker.MultipleLocator(base=200000))

plt.show()
```

```python


df_f['Specialization'].value_counts().plot(kind='barh', figsize=(20, 10))
plt.show()
```

```python

df_f_specialization_counts = df_f['Specialization'].value_counts()
df_m_specialization_counts = df_m['Specialization'].value_counts()
```

```python

# Create a list of specializations
specializations = df_m_specialization_counts.index.tolist()


fspecializations = df_f_specialization_counts.index.tolist()
# Create a list of female counts
female_counts = df_f_specialization_counts.values.tolist()

# Create a list of male counts
male_counts = df_m_specialization_counts.values.tolist()

# Create a figure and axes
fig, ax = plt.subplots(figsize=(20, 10))

# Create a horizontal bar chart for male counts
ax.barh(specializations, male_counts, label='Male')

# Create a horizontal bar chart for female counts
ax.barh(fspecializations, female_counts, label='Female')



# Add title and labels
ax.set_title('Specialization Counts by Gender', fontsize=16)
ax.set_xlabel('Number of Graduates', fontsize=14)
ax.set_ylabel('Specialization', fontsize=14)

# Add legend
ax.legend(fontsize=12)

# Show the plot
plt.show()
```
